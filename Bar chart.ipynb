```python
import pandas as pd
data = pd.read_excel('Data.xlsx')

vr_fe_columns = data.filter(regex='VR_FE_trial_\d+')
vr_wm_columns = data.filter(regex='VR_WM_trial_\d+')
vr_ltm_columns = data.filter(regex='VR_LTM_trial_\d+')
vr_pd1_columns = data.filter(regex='VR_PD1_trial_\d+')
vr_pd2_columns = data.filter(regex='VR_PD2_trial_\d+')
# Calculate the mean of each trial
data['Mean_VR_FE'] = vr_fe_columns.mean(axis=1)
data['Mean_VR_WM'] = vr_wm_columns.mean(axis=1)
data['Mean_VR_LTM'] = vr_ltm_columns.mean(axis=1)
data['Mean_VR_PD1'] = vr_pd1_columns.mean(axis=1)
data['Mean_VR_PD2'] = vr_pd2_columns.mean(axis=1)

ts_fe_columns = data.filter(regex='TS_FE_trial_\d+')
ts_wm_columns = data.filter(regex='TS_WM_trial_\d+')
ts_ltm_columns = data.filter(regex='TS_LTM_trial_\d+')
ts_pd1_columns = data.filter(regex='TS_PD1_trial_\d+')
ts_pd2_columns = data.filter(regex='TS_PD2_trial_\d+')
# Calculate the mean of each trial
data['Mean_TS_FE'] = ts_fe_columns.mean(axis=1)
data['Mean_TS_WM'] = ts_wm_columns.mean(axis=1)
data['Mean_TS_LTM'] = ts_ltm_columns.mean(axis=1)
data['Mean_TS_PD1'] = ts_pd1_columns.mean(axis=1)
data['Mean_TS_PD2'] = ts_pd2_columns.mean(axis=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the combinations of FL_group and MTL_group
combinations = ["Low FL High MTL", "Low MTL High FL", "Low FL Low MTL", "High FL High MTL"]

# Define the foraging tasks you want to plot
foraging_tasks = ['Mean_VR_FE', 'Mean_VR_WM', 'Mean_VR_LTM', 'Mean_VR_PD1', 'Mean_VR_PD2', 'Mean_TS_FE', 'Mean_TS_WM', 'Mean_TS_LTM', 'Mean_TS_PD1', 'Mean_TS_PD2']

# Initialize an empty dictionary to store the means for each combination
means = {combo: [] for combo in combinations}

# Calculate the means for each combination and each foraging task
for combo in combinations:
    for task in foraging_tasks:
        mean_value = data[data['FL_group'] + ' and ' + data['MTL_group'] == combo][task].mean()
        means[combo].append(mean_value)

# Convert means dictionary to a DataFrame
means_df = pd.DataFrame(means, index=foraging_tasks)

# Create the bar chart
means_df.plot(kind='bar', figsize=(12, 6))
plt.title('Mean Trial Scores for Foraging Tasks by FL_group and MTL_group')
plt.xlabel('Foraging Tasks')
plt.ylabel('Mean Trial Score')
plt.legend(title='FL_group and MTL_group Combinations')
plt.xticks(rotation=45)
plt.show()
```
